```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import math 
```

```python
temp_data = pd.read_csv("temp-data.csv",delimiter=";" ) # Read data 

'''
In one day there were more than one temp data but enocder and current data is daily 
So take avg of temp data in a day
'''

temp_data = (temp_data.groupby(["ROBO_ID","TEMP_DATE"]).mean()) 

temp_data.to_csv("temp-data.csv") # save this version 
```

```python
encoder_data = pd.read_csv("enkoder.csv",delimiter= ";")
encoder_data = encoder_data.drop(columns="TIME")
```

```python
encoder_data
```

```python
'''
Join tempeature and encoder data on ROBO ID and REG date 
'''
df_cd = pd.merge(temp_data, encoder_data, how='inner', left_on = ['ROBO_ID','TEMP_DATE'], right_on = ['ROBO_ID','REG_DATE']) 

'''
Rename joint columns to "jtname_e"
'''
df_cd = df_cd.rename(columns= {"JT1" : "JT1_e","JT2" : "JT2_e","JT3" : "JT3_e", "JT4" : "JT4_e","JT5" : "JT5_e",
                    "JT6" : "JT6_e","JT7" : "JT7_e","JT8" : "JT8_e"})
```

```python
data_cur = pd.read_csv("current-avg.csv",delimiter=';') # read_current_data

'''
Rename joint columns to "jtname_c"
'''
data_cur = data_cur.rename(columns= {"JT1" : "JT1_c","JT2" : "JT2_c","JT3" : "JT3_c", "JT4" : "JT4_c","JT5" : "JT5_c",
                    "JT6" : "JT6_c","JT7" : "JT7_c","JT8" : "JT8_c"})
```

```python

data_final = pd.merge(df_cd, data_cur, how='inner', left_on = ['ROBO_ID','REG_DATE'], right_on = ['ROBO_ID','REG_DATE']) #join current data too

data_final.to_csv("final-no-error.csv") # save final version 
```

```python
error_data = pd.read_csv("ERR_HIST.csv",delimiter=';') # read csv of error history

'''
Drop columns that Unnamed
'''
error_data = error_data.drop(['Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9',
       'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13',
       'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'],axis=1)
```
