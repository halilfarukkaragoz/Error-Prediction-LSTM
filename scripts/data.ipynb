### READ Final CSV file and stacked it for feed LSTM network 

```python
data = pd.read_csv("data_f_error.csv",delimiter=';')
```

```python
data = data[['ROBO_ID', 'REG_DATE','TEMP_VALUE', 'JT1_e', 'JT2_e', 'JT3_e', 'JT4_e',
       'JT5_e', 'JT6_e', 'JT7_e', 'JT8_e', 'JT1_c', 'JT2_c', 'JT3_c', 'JT4_c',
       'JT5_c', 'JT6_c', 'JT7_c', 'JT8_c', 'CODE']]
```

```python
data
```

```python
data.drop_duplicates()
```

```python
data.fillna(0,inplace=True)
```

```python
data
```

```python
error = data[data.CODE != 0]
```

```python
error_series = []
'''
Take the error periods 5 days before error 
'''

for row in error.iterrows():
   booleans = np.logical_and((np.logical_and(data.REG_DATE < row[1].REG_DATE, data.REG_DATE > row[1].REG_DATE-timedelta(days=10))), data['ROBO_ID'] == row[1]['ROBO_ID'])
   error_series.append([data[booleans].drop(columns=['REG_DATE']).to_numpy(),1])
   
```

```python
for count,value in enumerate(error_series):
    for count2,value2 in enumerate(value[0]):
        if value2[18] != 0:
            print(value2[18])
```

```python
error_series 
```

```python
non_error_series = []

for index,row in data.iterrows():
    counter = 0
    if(index %5 == 0):
        boolean1 = np.logical_and(data.CODE == 0 ,data['ROBO_ID'] == row['ROBO_ID'])
        boolean2 = np.logical_and(data.REG_DATE < row.REG_DATE, data.REG_DATE > row.REG_DATE-timedelta(days=6))
        boolean = np.logical_and(boolean1,boolean2)
        if np.sum(boolean) == 5:
            non_error_series.append([data[boolean].iloc[0:4].to_numpy(),0])
            counter+=1
            if counter == 2000:
                break
    
```

```python
error_array = np.array(error_series)
non_array = np.array(non_error_series)
```

```python
all_data = np.concatenate([error_array,non_array],axis=0)
np.save('data.npy',all_data)
```
